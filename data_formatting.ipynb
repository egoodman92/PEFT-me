```python
filename = "message_2"
data_loc = f"/Users/emmettgoodman/Documents/projects/peft-chatbot/messages/inbox/jennyquang_10156267310414458/{filename}.json"
```

```python
import json

with open(data_loc) as file:
    # Load the JSON data into a dictionary
    data = json.load(file)
```

```python
print(data["messages"])
print(len(data["messages"]))
messages = data["messages"][::-1]
print(messages[:5])
```

```python
dialogues = [["Noone", "Nothing"]]
for message in messages:
    if "content" in message:
        sender_name = message["sender_name"]
        content = message["content"]
        content = content.encode('latin1').decode('utf8')
    if sender_name == dialogues[-1][0]:
        dialogues[-1][1] += " \n "
        dialogues[-1][1] += content
    else:
        dialogues.append([sender_name, content])

dialogues = dialogues[1:]
if dialogues[0] == "Emmett Goodman":
    del dialogues[0]
```

```python
for_df = []

for i in range(0, len(dialogues) - 1, 2):
    for_df.append([f"Jenny Quang: {dialogues[i][1]} \n\n Emmett Goodman: {dialogues[i+1][1]}", "Johnny Appleseed", ["happy", "hungry"]])
```

```python
print(len(for_df))

import pandas as pd

df = pd.DataFrame(for_df, columns = ["quote", "author", "tags"])

df.to_json(f"jenny-emmett-{filename}.jsonl", orient='records', lines=True)
```
